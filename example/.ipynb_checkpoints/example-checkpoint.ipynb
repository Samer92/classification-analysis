```python
from classification_analysis.classification_analysis import MetricsVisualization, MetricsClaculation, ClassificationAnalysis

from sklearn.model_selection import train_test_split
from sklearn import datasets, metrics
from sklearn.svm import SVC

import numpy as np

import matplotlib.pyplot as plt
import pickle as pk

from pprint import pprint
```

<h2>Binary Classification </h2>

```python
# Prepare dataset
diabetes = datasets.load_breast_cancer()
X = diabetes.data
y = diabetes.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Feed SVM classifier
classifier = SVC(C=1.0, kernel='rbf', max_iter=-1)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
metrics.accuracy_score(y_test, y_pred)
```

```python
# plot confusion matrix
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
ax = plt.subplot(1,2,1)
MetricsVisualization.plot_confusion_matrix(confusion_matrix, list(diabetes.target_names), ax)
```

```python
# plot confusion table sample
ax = plt.subplot(1,2,1)
MetricsVisualization.plot_confusion_table_sample(ax)
```

<h2>Multiclass Classification </h2>

```python
# Prepare dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Feed SVM classifier
classifier = SVC(C=1.0, kernel='rbf', max_iter=-1)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
metrics.accuracy_score(y_test, y_pred)
```

```python
# plot confusion matrix
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
ax = plt.subplot(1,2,1)
MetricsVisualization.plot_confusion_matrix(confusion_matrix, list(diabetes.target_names), ax)
```
